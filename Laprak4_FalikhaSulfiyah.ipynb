```python
nilai = int(input("masukkan nilai : "))
if nilai >= 90:
    print("A")
elif nilai >= 80:
    print("B+")
elif nilai >= 70:
    print("B")
elif nilai >= 60:
    print("C+")
elif nilai >= 50:
    print("C")
elif nilai >= 40:
    print("D")
else:
    print("E")
```
